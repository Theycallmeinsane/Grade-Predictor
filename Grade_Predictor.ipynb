```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import matplotlib.pyplot as plt


```

Loading Data Set

```python
df = pd.read_csv('D:/Data1_copy_copy.csv')
scaler_df = MinMaxScaler()

```

```python
X1 = df[['Subject1', 'Subject2', 'Subject3', 'Subject4', 'Subject5', 'Subject6', 'Subject7']]
y1= df['Sem2%age']
```

Semester 2 Percentage Prediction

```python

X1 = df[['Subject1', 'Subject2', 'Subject3', 'Subject4', 'Subject5', 'Subject6', 'Subject7']]
y1= df['Sem2%age']

imputer = SimpleImputer(strategy='mean')
X1 = imputer.fit_transform(X1)



scaler = MinMaxScaler()
X1 = scaler.fit_transform(X1)
y1=scaler.fit_transform(y1.values.reshape(-1,1))

train_X1, test_X1, train_y1, test_y1 = train_test_split(X1, y1, test_size=0.2, random_state=7)
n_features1=train_X1.shape[1]

model1 = Sequential()
model1.add(Dense(50, input_dim=n_features1, activation='PReLU'))
model1.add(Dense(35, activation='PReLU'))
model1.add(Dense(25, activation='PReLU'))
model1.add(Dense(15, activation='PReLU'))

model1.add(Dense(1))

model1.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])

print(model1.summary())

model1.fit(train_X1 ,train_y1, epochs=300, batch_size=32)

y_pred=model1.predict(test_X1)
y_pred = y_pred.reshape(-1,)
test_y1=test_y1.reshape(-1,)

df_result=pd.DataFrame({'Actual':test_y1, 'Predicted':y_pred})
mse=mean_squared_error(test_y1, y_pred)
print(df_result)
 
print('MSE: ', mse)

```

```python
plt.figure(figsize=(8, 8))
plt.scatter(test_y1, y_pred, color='blue')
plt.plot([min(test_y1), max(test_y1)], [min(test_y1), max(test_y1)], linestyle='--', color='red', linewidth=2)
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.grid(True)
plt.show()
```
